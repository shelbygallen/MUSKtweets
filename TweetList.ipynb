```python
for x in range(len(tweets)-1, -1, -1):
    currTweet = tweets.iloc[x,0]
    if ((currTweet[0]+currTweet[1]) == "RT"):
        tweets = tweets.drop(tweets.index[x])
```

```python
# Set size of tweets to small(s), medium(m), or large(l)
# small = 1-47 characters
# medium = 48-93 characters
# large = 94-140 characters

for i in range(len(tweets)):  
    if len(tweets.iloc[i,0]) < 48:
        tweets.iloc[i,2] = 's'
    if len(tweets.iloc[i,0]) >= 48:
        if len(tweets.iloc[i,0]) <= 93:
            tweets.iloc[i,2] = 'm'
    if len(tweets.iloc[i,0]) > 93:
        tweets.iloc[i,2] = 'l'
tweets
    
```
